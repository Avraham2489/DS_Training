#### 88. How to implement the Game of Life using numpy arrays? (★★★)

````
board= np.random.randint(10,size=[10,10])>6
iter_number=10
for i in range(iter_number):
    windows=np.lib.stride_tricks.sliding_window_view(board,[3,3])
    alive_neigbours=np.sum(windows,axis=(2,3))-windows[:,:,1,1]
    alive_neigbours=np.pad(alive_neigbours,1)
    board[alive_neigbours<2]=0
    board[alive_neigbours==3]=1
    board[alive_neigbours>3]=0
    print('_______________________')
    print(board)
````
